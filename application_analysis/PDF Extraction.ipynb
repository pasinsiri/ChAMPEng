```python
filename = 'freshgrad_jobseek_2018.pdf'
```

```python
reader = PyPDF2.PdfFileReader(filename)
n = reader.getNumPages()

df = []
for page in [str(i+1) for i in range(n)]:
    df.append(tabula.read_pdf(filename, pages = page))
```

```python
# Use only table containing Bachelor's degree number (the first two)

# The first
df[0].head()
```

```python
# Rename columns
df[0].columns = ['Department', 'Unused_1', 'Working', 
                 'Working and Study', 'Total Working', 'Vacant', 
                 'Study', 'Filled In Grads','Total Grads', 
                 'Filled In Portion']

df[0].drop(['Unused_1'], axis = 1, inplace = True)

df[0] = df[0].iloc[[((i + 1) * 2) for i in range(1, 12)]]
```

```python
df[0].reset_index(drop = True, inplace = True)
df[0].head()
```

```python
df[1].head(20)
```

```python
df[1].columns = ['Unused_1', 'Department', 'Unused_2', 'Working', 
                 'Working and Study', 'Total Working', 'Vacant', 
                 'Study', 'Filled In Grads','Total Grads', 
                 'Filled In Portion']

df[1].drop(['Unused_1', 'Unused_2'], axis = 1, inplace = True)
df[1].head()
```

```python
#df[1] = df[1].iloc[[2,4,6,8,10,12]]
df[1].reset_index(drop = True, inplace = True)
df[1]
```

```python
# Concatenate two dataframes
final_table = pd.concat([df[0], df[1]], ignore_index=True)
final_table.head()
```

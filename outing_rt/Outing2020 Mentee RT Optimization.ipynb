```python
from pandas import DataFrame
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import pulp as plp
import random
```

```python
# 9 Rooms
path = 'Data/'
```

```python
temp_dset = pd.read_csv(path + 'input.csv')
temp_dset.drop_duplicates(subset = 'รหัสนิสิต', keep = 'last', inplace = True)
dset = temp_dset.drop(['Timestamp', 'ชื่อ', 'นามสกุล', 'ชื่อเล่น', 'Pair NO.'], axis = 1)
#dset.drop(columns='name',inplace=True)
```

```python
dset = dset.merge(dset,how='left',on='รหัสนิสิต',suffixes=['','2']).merge(dset,how='left',on='รหัสนิสิต',suffixes=['1','3'])
```

```python
dset.head(2)
```

```python
n = len(dset)
m = 27 #m = no. of rooms
set_I = range(0, n)
set_J = range(1, m+1)
```

```python
rank = {(i,j): dset.iloc[i,j] for i in set_I for j in set_J}
l = {(i,j): 0 for i in set_I for j in set_J}
u = {(i,j): 1 for i in set_I for j in set_J}
a = {(i,j): 1 for i in set_I for j in set_J}
b = {(j): 10 for j in set_J}
b2 = {(j): 5 for j in set_J}
b3 = {(i): 3 for i in set_I}
b4 = {(i): 1 for i in set_I}
b5 = {(i): 1 for i in set_I}
```

```python
#transform rank in to score
c = {}
for key in rank:
    if rank[key] == 1:
        c[key] = 16
    elif rank[key] == 2:
        c[key] = 8
    elif rank[key] == 3:
        c[key] = 4
    elif rank[key] == 4:
        c[key] = 2
    elif rank[key] == 5:
        c[key] = 1
    else:
        c[key] = -99
```

```python
opt_model = plp.LpProblem(name="MIP_Model")
```

```python
#assign lower bound and upper bound to decision variable
x_vars  = {(i,j):
plp.LpVariable(cat=plp.LpInteger,lowBound=l[i,j],upBound= u[i,j],name="x_{0}_{1}".format(i,j)) for i in set_I for j in set_J}
```

```python
#Room minimum and maximum attendee constraint
constraints = {j : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for i in set_I),sense=plp.LpConstraintLE,rhs=b[j],name="constraint_{0}".format(j)))for j in set_J}
constraints = {j : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for i in set_I),sense=plp.LpConstraintGE,rhs=b2[j],name="constraint2_{0}".format(j)))for j in set_J}
```

```python
#attendee must join 3 room
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in set_J),sense=plp.LpConstraintEQ,rhs=b3[i],name="constraint3_{0}".format(i)))for i in set_I}
#attendee can't join more than one room in the same time
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in set_J[0:9]),sense=plp.LpConstraintEQ,rhs=b4[i],name="constraint4_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in set_J[9:18]),sense=plp.LpConstraintEQ,rhs=b4[i],name="constraint5_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in set_J[18:27]),sense=plp.LpConstraintEQ,rhs=b4[i],name="constraint6_{0}".format(i)))for i in set_I}
#attendee must not join the same room
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [1,10,19]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint7_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [2,11,20]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint8_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [3,12,21]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint9_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [4,13,22]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint10_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [5,14,23]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint11_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [6,15,24]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint12_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [7,16,25]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint13_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [8,17,26]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint14_{0}".format(i)))for i in set_I}
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [9,18,27]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint15_{0}".format(i)))for i in set_I}
#constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(a[i,j] * x_vars[i,j] for j in [10,20,30]),sense=plp.LpConstraintLE,rhs=b5[i],name="constraint16_{0}".format(i)))for i in set_I}
```

```python
#minimum score (optimal == 28)
constraints = {i : opt_model.addConstraint(plp.LpConstraint(e=plp.lpSum(c[i,j] * x_vars[i,j] for j in set_J),sense=plp.LpConstraintGE,rhs=16,name="constraint17_{0}".format(i)))for i in set_I}
```

```python
objective = plp.lpSum(x_vars[i,j] * c[i,j] for i in set_I for j in set_J)
# for maximization
opt_model.sense = plp.LpMaximize
opt_model.setObjective(objective)
```

```python
opt_model.solve()
```

```python
summary = []
for i in set_I:
    for j in set_J:
        summary.append(plp.value(x_vars[i,j]*c[i,j]))
sum(summary)  
```

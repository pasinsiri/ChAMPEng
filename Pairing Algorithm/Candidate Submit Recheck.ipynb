```python
import pandas as pd
import numpy as np
```

# Prepare Recheck Table

```python
direct = pd.read_excel('application_score.xlsx', sheet_name='direct')
grey = pd.read_excel('application_score.xlsx', sheet_name='grey_score')
```

```python
direct_id = direct['รหัสนิสิต']
direct_id.dropna(inplace = True)
#direct_id
```

```python
final_table = grey[['รหัสนิสิต', 'Status (Y/G/N)']]
final_table.columns = ['ID', 'Status']
final_table.head()
```

```python
# Concat
temp_direct = pd.DataFrame(direct_id)
temp_direct.columns = ['ID']
temp_direct['Status'] = 'D'
```

```python
final_table = pd.concat([temp_direct, final_table], ignore_index=True)
final_table.head()
```

```python
# Convert status to score and save to csv
score_table = final_table.copy()
score_table['application_rank'] = final_table['Status'].map({
    'D': 4, 'Y': 3, 'GP': 2, 'GN': 1
})
score_table.drop(['Status'], axis = 1, inplace = True)
score_table
```
